```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Mid-Career 10th Percentile Salary'])
clean_df.insert(1, 'Spread', spread_col)
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread', ascending=True)
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
high_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
high_potential[['Undergraduate Major','Mid-Career 90th Percentile Salary']].head()
```

```python
max_spread_col = clean_df['Mid-Career 90th Percentile Salary'].subtract(clean_df['Starting Median Salary'])
clean_df.insert(5,"Max Spread", max_spread_col)
clean_df.head()
```

```python
clean_df.head()
```

```python
greatest_spread = clean_df.sort_values('Max Spread', ascending=False)
greatest_spread[['Undergraduate Major', 'Max Spread']].head()
```
